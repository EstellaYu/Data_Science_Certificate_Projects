# Share Ride Analysis
### Data: 
1. city data (`city name`, `number of drivers`, `city type`: Urban / Suburban / Rural)
2. ride data (`city name`, `ride date`, `fare`, `ride_id`)

### Analysis:
1. Create a `bubble plot`, displaying the relationship between the `total number of rides` in each city and the `average cost per ride`


2. `Pie charts` based on city type: 
    1. `total ride fares` 
    
    2. `total number of rides` 
    
    3. `total number of drivers`
    
    
3. A comparative `bar chart`, displaying `total ride fare` and `average earning per driver` in each city type 

### Conclusions
1. Of the `3` ciy types, the **car share bussiness is the `most active` in `Urban` cities**, with the `number of rides` is ranked as: `Urban` > `Suburban` > `Rural`; maybe it's due to the fact that the **`average fare` per trip is the least expensive in urban cities**. `Average fare` based on city type: `Urban` < `Suburban` < `Rural`.


2. Due to the popularity of car sharing in urban cities, the `total fare`, `total number of rides`, and the `total number of drivers` are all **increasing** from `rural`, `suburban` to `urban` cities. 
    * It's noticable that the percentage of `total fare`, `total number of rides` have similar values for the 3 city types. However, `Urban` cities have significantly **more drivers** than the other 2 types 
    
    * (which leads to the following analysis and conclution in the bar charts)
    
    
3. From the comparative bar charts, we can see an _**opposite trend**_ in `Total fare` and `Average earning per driver` based on different city types. The seeminly car sharing popularity in urban cities definitely brings bussiness to the car sharing company, but NOT neccessarily to individual drivers. 
    * due to the `small number of drivers` and `more expensive rides` in rural cities, driving in rural areas might financially benefit the individual drivers more 
    * So in terms of financially benefits: 
        * For car sharing company: `Urban` > `Suburban` > `Rural`
        * For individual drivers: `Rural` > `Suburban` > `Urban`

```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)
print(city_df.head())
print(ride_df.head())

# Combine the data into a single dataset
city_ride_df = pd.merge(ride_df, city_df, on = 'city', how = 'outer')

# Display the data table for preview
city_ride_df.head()
```

```python
print(city_ride_df.count())
print()
print("There is no empty entries in the outer merge.")
```

## Bubble Plot of Ride Sharing Data

```python
# find city types
city_type = list(city_ride_df.groupby("type").count().index)
print("There are {} types of cities: {}".format(len(city_type), 
                                                ", ".join(x for x in city_type[:-1]) + " and " + city_type[-1] +"."))
```

```python
# initiate the lists
cities_by_type = []               # city names by city type
total_ride_per_city = []          # total ride per city by city type
average_fare_per_city = []        # average fare per city by city type
driver_count_per_city = []

# loop through each type
for x in city_type:
    
    # create a "groupby" object in each city type
    sub_city_type_df = city_ride_df[city_ride_df["type"] == x]
    sub_city_group = sub_city_type_df.groupby("city")
    
    # obtain city name, total rides in each city, and average fare in each city
    cities = list(sub_city_group.first().index)
    sub_city_total_ride = list(sub_city_group.count()['ride_id'])
    sub_city_average_fare = list(sub_city_group.mean()["fare"])
    sub_city_driver_count = list(sub_city_group.first()['driver_count'])
    
    # append the list to the previously created list
    cities_by_type.append(cities)
    total_ride_per_city.append(sub_city_total_ride)
    average_fare_per_city.append(sub_city_average_fare)
    driver_count_per_city.append(sub_city_driver_count)
```

```python
# normalized driver_count_per_city
# min and max driver in all cities 
min_driver = city_ride_df['driver_count'].min()  # set to size 1
max_driver = city_ride_df['driver_count'].max()  # set to size 10

min_size = 10
max_size = 800

new_driver_count_per_city = [[(y-min_driver)/(max_driver - min_driver) * (max_size - min_size) + min_size
                              for y in x] 
                             for x in driver_count_per_city]
```

```python
# pre-set FaceColor
facecolor = ["Gold", "SkyBlue", "Coral"]

plt.figure(figsize = (9, 4))
# Build the scatter plots for each city types
for i in range(len(city_type)):
    plt.scatter(total_ride_per_city[i], average_fare_per_city[i], 
                s = new_driver_count_per_city[i],
                marker = "o",
                facecolors = facecolor[i],
                edgecolors = "black",
                alpha = 0.7,
               label = city_type[i])

# Incorporate the graph properties (axis_labels, title, grid)
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.grid(True)

# Create a legend
plt.legend(loc = "best", title = "City Types")

# Incorporate a text label regarding circle size
plt.text(43, 37, "Note:")
plt.text(43, 35, "Circle Size correlates with driver count per city.")

# Save Figure
plt.tight_layout()
plt.savefig("total_rides_vs_average_fare.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fare_per_type = city_ride_df.groupby("type").sum()["fare"]
total_fare_per_type_pcnt = total_fare_per_type / sum(total_fare_per_type)

# Build Pie Chart
plt.pie(total_fare_per_type_pcnt,
        radius = 1.5,
        labels = city_type,
        autopct = "%1.1f%%",
        explode = [0, 0, 0.1],
        colors = facecolor,
        shadow = True,
        startangle = 140, 
       )

# set figure properties
plt.title("% of Total fares by city type\n\n".title())

# Save Figure
plt.tight_layout()
plt.savefig("pct_of_total_fare_by_city_type.png")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
total_rides_city_type = city_ride_df.groupby("type").count()["ride_id"]
total_rides_city_type_pcnt = total_rides_city_type / total_rides_city_type.sum()

# Build Pie Chart
plt.pie(total_rides_city_type_pcnt,
       radius = 1.5,
        labels = city_type,
        autopct = "%1.1f%%",
        explode = [0, 0, 0.1],
        colors = facecolor,
        shadow = True,
        startangle = 140)

# set figure properties
plt.title("% of Total Rides by city type\n\n".title())

# Save Figure
plt.tight_layout()
plt.savefig("pct_of_total_rides_by_city_type.png")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_city_type_df = city_ride_df.groupby("city").first()[["driver_count","type"]]
total_driver_city_type = driver_city_type_df.groupby("type").sum()["driver_count"]
total_driver_city_type_pcnt = total_driver_city_type / total_driver_city_type.sum()

# Build Pie Chart
plt.pie(total_driver_city_type_pcnt,
       radius = 1.5,
        labels = city_type,
        autopct = "%1.1f%%",
        explode = [0, 0, 0.1],
        colors = facecolor,
        shadow = True,
        startangle = 140)

# set figure properties
plt.title("% of Total Drivers by city type\n\n".title())

# Save Figure
plt.tight_layout()
plt.savefig("pct_of_total_drivers_by_city_type.png")

# Show Figure
plt.show()
```
